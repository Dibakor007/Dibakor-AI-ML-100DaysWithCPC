```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'Hours_study':[1,2,3,4,5],
    'Hours_phone':[1,2,4,6,7],
    'Score':[38,50,78,94,80]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
x = df[['Hours_study','Hours_phone']]
y = df['Score']
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
p_score = model.predict([[4,6]])
```

```python
p_score
```
