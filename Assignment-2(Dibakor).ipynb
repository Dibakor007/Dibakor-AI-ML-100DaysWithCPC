```python
df['sex'] = df['sex'].fillna('N')
```

```python
df['sex'].isnull().sum()
```

```python
df['TSH'] = df['TSH'].fillna(df['TSH'].mean())
```

```python
df['T3'] = df['T3'].fillna(df['T3'].median())
```

```python
df['TT4'] = df['TT4'].fillna(df['TT4'].median())
```

```python
df['T4U'] = df['T4U'].fillna(df['T4U'].median())
df['FTI'] = df['FTI'].fillna(df['FTI'].mean())
df['TBG'] = df['TBG'].fillna(df['TBG'].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Le = LabelEncoder()
```

```python
df['sex'] = Le.fit_transform(df['sex'])
```

```python
for col in df.select_dtypes(include=['object']).columns:
    df[col] = Le.fit_transform(df[col].astype(str))
```
